# Step 1: Feature Engineering

### Objectives:
We'll address the following:

1. **Group Sparse Categories**  
   - Combine infrequent categories into broader groups (e.g., in `Status`, `CreditHistory`, `Purpose`).

2. **Engineer Risk Scores**  
   - Assign risk scores to categories based on their proportion of Bad credit risks (`CreditRisk = 2`).

3. **Create Interaction Terms**  
   - Generate combinations of categorical features (e.g., `Status_CreditHistory`) to capture interaction effects.

---




```python
# Define features (X) and target (y)
X = data.drop("CreditRisk", axis=1)
y = data["CreditRisk"]

# Verify the structure of X and y
print("Feature Matrix (X):")
print(X.head())
print("\nTarget Variable (y):")
print(y.head())

```

### Confirm the Shape and Columns of X
After defining X, it’s always a good idea to check its structure to ensure all features are included:

```python
print("Shape of X:", X.shape)
print("Columns in X:", X.columns)
```

#### Group Sparse Categories
Simplified less frequent categories in Status, CreditHistory, and Purpose.

```python
## Group Sparse Categories
# Combine infrequent categories into broader groups
X["Status"] = X["Status"].replace({"A13": "Other"})
X["CreditHistory"] = X["CreditHistory"].replace({"A30": "Other", "A31": "Other"})
X["Purpose"] = X["Purpose"].replace({"A48": "Other", "A49": "Other"})
```

#### Engineer Risk Scores
Created StatusRiskScore and CreditHistoryRiskScore based on categorical feature risk mappings.

```python
# risk mapping with a default score
risk_mapping_status = {
    "A11": 0.7,
    "A12": 0.4,
    "A13": 0.2,
    "A14": 0.1,
    "Other": 0.5  # Default risk score for unaccounted categories
}

# Apply the mapping again
X["StatusRiskScore"] = X["Status"].map(risk_mapping_status)

# Fill any remaining NaN values with the default risk score
X["StatusRiskScore"].fillna(0.5, inplace=True)

# Verify that missing values are resolved
print("Missing values in StatusRiskScore after fix:", X["StatusRiskScore"].isnull().sum())
```

```python
# Check columns after creating StatusRiskScore
print("Columns in the dataset after creating StatusRiskScore:", X.columns)

# Check if any of the required columns are missing
required_columns = {'Status', 'CreditHistory', 'Purpose', 'Status_CreditHistory'}
missing_columns = required_columns - set(X.columns)
if missing_columns:
    print(f"Missing columns: {missing_columns}")
else:
    print("All required columns are present.")
```

```python
print("Columns in the dataset at this step:", X.columns)
```

## What Happened?

### Root Cause of Missing Values:
- The **Status** column had categories that were not included in the original `risk_mapping_status` dictionary.
- When using `.map()` to assign scores, any category not found in the dictionary was mapped to `NaN`.

---

### Fix Applied:
- The `risk_mapping_status` dictionary was updated to include:
  - All known categories from the **Status** column.
  - A default risk score (`0.5`) for any unaccounted categories, ensuring no `NaN` values were introduced.

---

### Filled Missing Values:
- After applying the mapping, any remaining `NaN` values in **StatusRiskScore** were filled with a default score (`0.5`) using `.fillna()`.

# Step 2: Incorporate Adjustments into Preprocessing Pipeline

## Objective
Integrate all preprocessing steps into a single pipeline:
1. **One-Hot Encoding:** Encode categorical variables (including the interaction term `Status_CreditHistory`).

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Define numerical and categorical features
numerical_features = ["Duration", "CreditAmount", "StatusRiskScore", "CreditHistoryRiskScore", "Age"]
categorical_features = ["Status", "CreditHistory", "Purpose", "Status_CreditHistory"]

# Create a preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numerical_features),
        ("cat", OneHotEncoder(drop="first"), categorical_features),
    ]
)

# Apply preprocessing pipeline
X_transformed = preprocessor.fit_transform(X)

# Check the transformed data shape
print("Transformed Data Shape:", X_transformed.shape)
```

# Step 4: Build a Baseline Model

Now that the data is split into training and testing sets, we will build a baseline model to:

- **Establish a reference point** for evaluating future, more complex models.
- Use a **simple, interpretable algorithm**, such as Logistic Regression, to obtain initial results.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Initialize the Logistic Regression model
baseline_model = LogisticRegression(class_weight="balanced", random_state=42, max_iter=500)

# Train the model on the training set
baseline_model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = baseline_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label=2)  # Bad CreditRisk = 2
recall = recall_score(y_test, y_pred, pos_label=2)
f1 = f1_score(y_test, y_pred, pos_label=2)

# Display metrics
print(f"Baseline Model Performance:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# Display a confusion matrix and classification report
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from scipy.stats import ks_2samp

def detect_drift(training_data, new_data, feature_columns, threshold=0.05):
    """
    Detect data drift using KS statistic for numerical features.

    Args:
        training_data (pd.DataFrame): Original training dataset.
        new_data (pd.DataFrame): New incoming dataset to compare.
        feature_columns (list): List of numerical feature columns.
        threshold (float): P-value threshold for detecting drift.

    Returns:
        pd.DataFrame: Drift results including KS statistic, p-value, and drift detection flag.
    """
    drift_results = []
    for feature in feature_columns:
        ks_stat, p_value = ks_2samp(training_data[feature], new_data[feature])
        drift_results.append({
            "Feature": feature,
            "KS Statistic": ks_stat,
            "P-Value": p_value,
            "Drift Detected": p_value < threshold
        })

    return pd.DataFrame(drift_results)

# Example usage with test data
drift_results = detect_drift(X_train, X_test, numerical_features)
print("Drift Detection Results:")
print(drift_results)
```
